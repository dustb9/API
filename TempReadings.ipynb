```python
import matplotlib.pyplot as plt
from matplotlib import style
import pandas as pd
import numpy as np
import datetime as dt
from os.path import isfile,exists,splitext,join,isdir
import glob
from scipy.signal import savgol_filter
import datetime
"import pylibsie as sie"
import seaborn as sb
import shutil
style.use('default')
#from tqdm import tqdm
```

```python

pro = "306-6"

AccumTemp = []
AmbTemp = []
ValveBlockTemp = []
LHCylTemp = []

filepath = r'I:\308\Field Data\2022\01_DAILY_FILES\\' + pro + r'\\*.csv' 
for file in glob.glob(filepath):
    if "OBS" in file:
        df = pd.read_csv(file)
        if "308_ACCUM_TEMP [C]" in df.columns:
            AccumTemp.extend(df['308_ACCUM_TEMP [C]'].tolist())
            AmbTemp.extend(df['308_AMB_TEMP_C [C]'].tolist())
            ValveBlockTemp.extend(df['308_VALVE_BLOCK_TEMP [c]'].tolist())
            LHCylTemp.extend(df['LH_CYL_TEMP_C [C]'].tolist())
            print(file)
        else:
            pass
```

```python
print(len(AccumTemp))
print(len(AmbTemp))
print(len(ValveBlockTemp))
print(len(LHCylTemp))
```

```python
df = pd.DataFrame({"AccumTemp":AccumTemp, "AmbTemp":AmbTemp, "ValveBlockTemp":ValveBlockTemp, "LHCylTemp":LHCylTemp})
df = df.query('AmbTemp < 3000')
```

```python
df.plot(y = "AccumTemp", title = "AccumTemp_" + pro + "_June-Oct", ylabel="C")
print(df['AccumTemp'].max())
print(df['AccumTemp'].min())
print(df['AccumTemp'].mean())
```
